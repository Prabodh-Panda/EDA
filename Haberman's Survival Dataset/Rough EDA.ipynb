```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
habermans = pd.read_csv("Haberman's Dataset.csv")
habermans.head()
```

## Features
1. Age - Age of patient
2. Year of Operation - Year of operation (1900)
3. Axillary Nodes - Lymph Nodes
4. Survival Class - 1 (Survived 5 Years or Longer)
                    2 (Survived less than 5 years)

```python
habermans['survival_class'].value_counts()
```

```python
sns.pairplot(habermans, hue='survival_class', kind='scatter')
```

```python
habermans_class_1 = habermans[habermans['survival_class'] == 1]
habermans_class_2 = habermans[habermans['survival_class'] == 2]
```

```python
sns.FacetGrid(habermans,hue='survival_class').map(sns.histplot,'age', kde=True, stat='density', kde_kws=dict(cut=3)).add_legend()
plt.title("Histogram with PDF for 'age' feature")
```

```python
sns.FacetGrid(habermans,hue='survival_class').map(sns.histplot,'year_of_operation', kde=True, stat='density', kde_kws=dict(cut=3)).add_legend()
plt.title("Histogram with PDF for 'year_of_operation' feature")
```

```python
sns.FacetGrid(habermans,hue='survival_class').map(sns.histplot,'axillary_nodes', kde=True, stat='density', kde_kws=dict(cut=3)).add_legend()
plt.title("Histogram with PDF for 'axillary_nodes' feature")
```

```python
count,bin_edges = np.histogram(habermans['axillary_nodes'],density=True,bins=10)
PDF = count/sum(count)
CDF = np.cumsum(PDF)
print(CDF)
```

```python
plt.plot(bin_edges[1:],PDF,label = "PDF",)
plt.plot(bin_edges[1:],CDF,label = "CDF")
plt.legend()
plt.xlabel("axillary_nodes")
plt.ylabel("probability")
plt.title("PDF and CDF of 'axillary_nodes' feature")
```

```python
#plot for 'survived>=5yrs'
count,bin_edges = np.histogram(habermans_class_1['axillary_nodes'],bins=20,density=True)
PDF = count/sum(count)
#calculating CDF
CDF = np.cumsum(PDF)
plt.plot(bin_edges[1:],PDF,label="PDF of survived>=5yrs")
plt.plot(bin_edges[1:],CDF,label="CDF of survived>=5yrs")

#plot for 'survived<5yrs'
count,bin_edges = np.histogram(habermans_class_2['axillary_nodes'],bins=20,density=True)
PDF = count/sum(count)
#calculating CDF
CDF = np.cumsum(PDF)
plt.plot(bin_edges[1:],PDF,label="PDF of survived<5yrs")
plt.plot(bin_edges[1:],CDF,label="CDF of survived<5yrs")
plt.legend()
plt.xlabel("axillary_nodes")
plt.ylabel("probability")
plt.title("PDF and CDF of both survival_classes")
```
